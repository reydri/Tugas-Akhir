```python
import cv2
import numpy as np
from skimage import io
from PIL import Image
import pandas as pd
import glob
import os
from matplotlib import pyplot as plt
from pprint import pprint as pp

path_in = 'crop/*'

size_max_image = 500
debug_mode = True

data = pd.read_csv('datasetkeliling.csv')
def show_image(image, window_name):
    # Show image
    cv2.namedWindow(window_name, cv2.WINDOW_NORMAL)
    cv2.imshow(window_name, image)
    image_width, image_height = get_image_width_height(image)
    cv2.resizeWindow(window_name, image_width, image_height)

    # Wait before closing
    cv2.waitKey(0)
    cv2.destroyAllWindows()
    
def get_image_width_height(image):
    image_width = image.shape[1]  # current image's width
    image_height = image.shape[0]  # current image's height
    return image_width, image_height

```

```python
keliling = []
Xo =[]
yo = []
xup=[]
xdn=[]
xrg=[]
xlf=[]
cb = 0
for file_iterator in glob.iglob(path_in):
    
    debug_mode = True
    image = cv2.imread(file_iterator)
    image_yuv = cv2.cvtColor(image, cv2.COLOR_BGR2YUV)
    image_y = np.zeros(image_yuv.shape[0:2], np.uint8)
    image_y[:, :] = image_yuv[:, :, 0]

    image_blurred = cv2.GaussianBlur(image_y, (3, 3), 0)


    # canny edge-detector
    edges = cv2.Canny(image_blurred, 100, 300, apertureSize=3)
    #if (debug_mode):  show_image(edges, window_name)
    
    # keliling
    kel = 0
    for i in range(len(edges)):
        for j in range(len(edges[1])):
            if edges[i][j] == 255:
                kel += 1
    keliling.append(kel)
    
    # d1 d2
    M = cv2.moments(edges)
    
    X0 = int(M["m10"] / M["m00"])
    Y0 = int(M["m01"] / M["m00"])
    X1 = X0 + 1
    Y1 = Y0
    
    #print(X0,Y0)
    Xo.append(X0)
    yo.append(Y0)
    rgb = cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
    cv2.circle(rgb, (X0, Y0), 3, (255,255,255),-1)
    #if (debug_mode):  show_image(rgb, window_name)
    cb +=1
    
    # X makin besar semakin ke bawah
    # X ke bawah
    lngth = 0
    count = 0
    for i in range(X0, len(edges), 1):
        if edges[i][Y0] == 0:
            lngth += 1
        else:
            count = lngth + 1
            lngth += 1
    if count == 0 or count < 8:
        count = lngth +1
    xdn.append(count)
    
    # X keatas
    lngth = 0
    count = 0
    for i in range(X0, -Y0, -1):
        if edges[i][Y0] == 0:
            lngth += 1
        else:
            count = lngth + 1
            lngth += 1
    if count == 0 or count < 8:
        count = lngth +1
    xup.append(count)

    # Y kekanan
    lngth = 0
    count = 0
    for i in range(Y0, len(edges[X0]), 1):
        
            
        if edges[X0][i] == 0:
            lngth += 1
        else:
            count = lngth + 1
            lngth += 1
    if count == 0 or count < 8:
        count = lngth +1
        
    
    xrg.append(count)    
       
    #Y kekiri
    lngth = 0
    count = 0
    for i in range(X0, -Y0, -1):
        if edges[X0][i] == 0:
            lngth += 1
        else:
            count = lngth + 1
            lngth += 1
    if count == 0 or count < 8:
        count = lngth +1
    xlf.append(count)

    
# menghitung d1 dan d2        
d1 = []
d2 = []
for i in range(100):
    d1.append(xup[i] + xdn[i])
for i in range(100):
    d2.append(xrg[i] + xlf[i])
df = {'d1':d1,'d2':d2}

df =pd.DataFrame(df)
```

```python
# menghitung D

D= []
for i in range(len(d1)):
    D.append((d1[i]+d2[i])/2)

# menghitung A
A=[]
phi = 22/7
for i in range(len(D)):
    A.append((1/4)*phi*(D[i]**2))
A
```
