```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import cv2
import numpy as np
from matplotlib import colors
```

```python
img = mpimg.imread('in/b1.jpg')
```

```python
img.shape
```

```python
imgplot = plt.imshow(img) #show picture
```

```python
img_cropped = img[450:570, 460:580]
plt.imshow(img_cropped) #show image croppedx
```

```python
#Convert BGR TO RGB
RGB_im = cv2.cvtColor(img_cropped, cv2.COLOR_BGR2GRAY)
RGB_im = cv2.resize(img_cropped, None, fx=1/3, fy=1/3)
```

```python
plt.imshow(RGB_im)
plt.title('RGB')
plt.show()
```

```python
print('Type of the image : ' , type(RGB_im))
print()
print('Shape of the image : {}'.format(RGB_im.shape))
print('Image Height {}'.format(RGB_im.shape[0]))
print('Image Width {}'.format(RGB_im.shape[1]))
print('Dimension of Image {}'.format(RGB_im.ndim))
```

```python
print('Image size {}'.format(RGB_im.size))
print('Maximum RGB value in this image {}'.format(RGB_im.max()))
print('Minimum RGB value in this image {}'.format(RGB_im.min()))
```

```python
#Let's pick a specific pixel located at 100 th Rows and 50 th Column. 
#And view the RGB value gradually. 
RGB_im[ 25, 30 ]
```

```python
# A specific pixel located at Row : 100 ; Column : 50 
# Each channel's value of it, gradually R , G , B
print('Value of only Red {}'.format(RGB_im[ 25, 30, 0]))
print('Value of only Greed {}'.format(RGB_im[ 25, 30, 1]))
print('Value of only Blue {}'.format(RGB_im[ 25, 30, 2]))
```

```python
plt.title('Red')
plt.ylabel('Height {}'.format(RGB_im.shape[0]))
plt.xlabel('Width {}'.format(RGB_im.shape[1]))

plt.imshow(RGB_im[ : , : , 0])
plt.show() #show image channel red from label height and width
```

```python
plt.title('Green')
plt.ylabel('Height {}'.format(RGB_im.shape[0]))
plt.xlabel('Width {}'.format(RGB_im.shape[1]))

plt.imshow(RGB_im[ : , : , 1])
plt.show()
```

```python
plt.title('Blue')
plt.ylabel('Height {}'.format(RGB_im.shape[0]))
plt.xlabel('Width {}'.format(RGB_im.shape[1]))

plt.imshow(RGB_im[ : , : , 2])
plt.show()
```

```python
RGB_im[390:400 , : , 0] = 255 # full intensity to those pixel's Red
plt.figure( figsize = (10,10))
plt.imshow(RGB_im)
plt.show() #show pixel from channel red
```

```python
RGB_im[410:420 , : , 1] = 255 # full intensity to those pixel's Green
plt.figure( figsize = (10,10))
plt.imshow(RGB_im)
plt.show()
```

```python
RGB_im[430:440 , : , 2] = 255 # full intensity to those pixel's Blue
plt.figure( figsize = (10,10))
plt.imshow(RGB_im)
plt.show()
```

```python
RGB_im[ 390:440 , 410:420 , [0,1,2] ] = 200 # full intensity to those pixel's white
plt.figure( figsize = (10,10))
plt.imshow(RGB_im)
plt.show()
```

```python
fig, ax = plt.subplots(nrows = 1, ncols=3, figsize=(15,5))

for c, ax in zip(range(3), ax):
    
    # create zero matrix
    split_img = np.zeros(RGB_im.shape, dtype="uint8") # 'dtype' by default: 'numpy.float64'
    
    # assing each channel 
    split_img[ :, :, c] = RGB_im[ :, :, c]
    
    # display each channel
    ax.imshow(split_img)
```
